```python
from PIL import Image #Importing necesssary library

#Defining all characters in a 5 cross 7 binary matrix
character = {
    "A": ["01110","10001","10001","11111","10001","10001","10001"],
    "B": ["11110","10001","10001","11110","10001","10001","11110"],
    "C": ["01110","10001","10000","10000","10000","10001","01110"],
    "D": ["11110","10001","10001","10001","10001","10001","11110"],
    "E": ["11111","10000","10000","11110","10000","10000","11111"],
    "F": ["11111","10000","10000","11110","10000","10000","10000"],
    "G": ["01110","10001","10000","10111","10001","10001","01110"],
    "H": ["10001","10001","10001","11111","10001","10001","10001"],
    "I": ["01110","00100","00100","00100","00100","00100","01110"],
    "J": ["00111","00010","00010","00010","10010","10010","01100"],
    "K": ["10001","10010","10100","11000","10100","10010","10001"],
    "L": ["10000","10000","10000","10000","10000","10000","11111"],
    "M": ["10001","11011","10101","10101","10001","10001","10001"],
    "N": ["10001","11001","10101","10011","10001","10001","10001"],
    "O": ["01110","10001","10001","10001","10001","10001","01110"],
    "P": ["11110","10001","10001","11110","10000","10000","10000"],
    "Q": ["01110","10001","10001","10001","10101","10010","01101"],
    "R": ["11110","10001","10001","11110","10100","10010","10001"],
    "S": ["01111","10000","10000","01110","00001","00001","11110"],
    "T": ["11111","00100","00100","00100","00100","00100","00100"],
    "U": ["10001","10001","10001","10001","10001","10001","01110"],
    "V": ["10001","10001","10001","10001","10001","01010","00100"],
    "W": ["10001","10001","10001","10101","10101","11011","10001"],
    "X": ["10001","10001","01010","00100","01010","10001","10001"],
    "Y": ["10001","10001","01010","00100","00100","00100","00100"],
    "Z": ["11111","00001","00010","00100","01000","10000","11111"],
    '0': ["01110","10001","10011","10101","11001","10001","01110"],
    '1': ["00100","01100","00100","00100","00100","00100","01110"],
    '2': ["01110","10001","00001","00110","01000","10000","11111"],
    '3': ["01110","10001","00001","00110","00001","10001","01110"],
    '4': ["00010","00110","01010","10010","11111","00010","00010"],
    '5': ["11111","10000","11110","00001","00001","10001","01110"],
    '6': ["00110","01000","10000","11110","10001","10001","01110"],
    '7': ["11111","00001","00010","00100","01000","01000","01000"],
    '8': ["01110","10001","10001","01110","10001","10001","01110"],
    '9': ["01110","10001","10001","01111","00001","00010","01100"],
    '(': ["00010","00100","01000","01000","01000","00100","00010"],
    ')': ["01000","00100","00010","00010","00010","00100","01000"],
    '[': ["00110","00100","00100","00100","00100","00100","00110"],
    ']': ["01100","00100","00100","00100","00100","00100","01100"],
    '{': ["00010","00100","00100","01000","00100","00100","00010"],
    '}': ["01000","00100","00100","00010","00100","00100","01000"],
    '/': ["00001","00010","00100","01000","10000","00000","00000"],
    '\\':["10000","01000","00100","00010","00001","00000","00000"],
     '+':["00000","00100","00100","11111","00100","00100","00000"],
    '*': ["00000","01010","00100","11111","00100","01010","00000"],
    '=': ["00000","00000","11111","00000","11111","00000","00000"],
    '<': ["00010","00100","01000","10000","01000","00100","00010"],
    '>': ["01000","00100","00010","00001","00010","00100","01000"],
    '@': ["01110","10001","10111","10101","10111","10000","01110"],
    '#': ["01010","11111","01010","11111","01010","11111","01010"],
    '$': ["01110","10101","10100","01110","00101","10101","01110"],
    '%': ["11001","11010","00100","01000","10011","00011","00000"],
    '&': ["01100","10010","10100","01000","10101","10010","01101"],
    ' ': ["00000","00000","00000","00000","00000","00000","00000"],
    '.': ["00000","00000","00000","00000","00000","00100","00100"],
    ',': ["00000","00000","00000","00000","00100","00100","01000"],
    '!': ["00100","00100","00100","00100","00100","00000","00100"],
    '?': ["01110","10001","00010","00100","00100","00000","00100"],
    ':': ["00000","00100","00100","00000","00100","00100","00000"],
    ';': ["00000","00100","00100","00000","00100","00100","01000"],
    '-': ["00000","00000","00000","11111","00000","00000","00000"],
    '_': ["00000","00000","00000","00000","00000","00000","11111"],
    "'": ["00100","00100","00000","00000","00000","00000","00000"],
    '"': ["01010","01010","00000","00000","00000","00000","00000"],
}

#If apart from these there is some extra character then using this question mark character in place of it
unknownchar = ["01110","10001","00001","00110","00100","00000","00100"]


def rendertexttoimage(   word: str, #input string which we want to render
                         scalepix: int = 8, #This tells us how large each pixel is
                         charspace: int = 1, #spacing between characters
                         linespace: int = 2, #Spacing between lines
                         paddingback: int = 8, #padding space around text written
                         backcolor: int = 0, #Color of background
                         forecolor: int = 255) : #color of foreground pixel

    line = word.split("\n") #spilts text into lines whenever we see \n
    line = [ln.upper() for ln in line] #converting all letters to capital letters

    charwidth = 5  #Each character in image is 5 pixel in width and 7 pixel in height
    charheight = 7

   #Calculating width of each line if its empty then width is 0
    linepixelwidth = []
    for ln in line:
        if len(ln) == 0:
            linepixelwidth.append(0)
            continue
        linepixelwidth.append(len(ln) * charwidth + (max(0, len(ln)-1) * charspace)) #calculating width

    maxlinewidthpixels = max(linepixelwidth) if linepixelwidth else 0 #using max funtion to determine widest line

    #Total pixel rows of characters
    totalrows = len(line) * charheight + max(0, len(line)-1) * linespace
    imagewidth = maxlinewidthpixels * scalepix + 2 * paddingback #calculating width and height of image
    imageheight = totalrows * scalepix + 2 * paddingback

    #Creating L mode of image and filling all pixels with background color
    image = Image.new('L', (imagewidth, imageheight), color=backcolor)
    pixelset = image.load() #this allows us to set each pixel individually

    characterrows = 0 #Keeping track of how many 7 pixels rows are there.Initially setting it to 0
    for lineindex, ln in enumerate(line): #looping through each line of text
        horizontalposition = 0  #resets horizontal position at start of each new line
        for charindex, ch in enumerate(ln): #loops through each character of current line
            bitmap = character.get(ch, unknownchar) #Getting of bitmap of character which we encounter while looping


            for row in range(charheight): #loops through each row  of bitmap
                stringrow = bitmap[row] #It holds string of 1 and 0's of that particular row of bitmap of character
                for col in range(charwidth): #loops through each column of row
                    bitvalue = stringrow[col] #gets bit value of that column of one row
                    if bitvalue == "1": #that is our character pixel
                        x0 = paddingback + (horizontalposition  + col) * scalepix #calculates horizontal pixel position of final image for top left corner
                        y0 = paddingback + (characterrows + row) * scalepix #Calculates vertical pixel position of final image for top left corner

                        #Looping vertically over the scaled size of a single character pixel.
                        for unscaledy in range(scalepix):
                            scaledy = y0 + unscaledy #Computing scaled y coordinate
                            for unscaledx in range(scalepix): #similarly computing scaled coordinate of pixel
                                scaledx = x0 + unscaledx

                               #to ensure that pixel coordinates are inside image bounds
                                if 0 <= scaledx < imagewidth and 0 <= scaledy < imageheight:

                                    pixelset[scaledx, scaledy] = forecolor #setting these pixels as foreground pixels
            #moving horizontally to next character
            horizontalposition += charwidth + charspace

        #after finishing horizontal line moving to next line
        characterrows += charheight + linespace

    return image #getting final image

if __name__ == "__main__":
    #sample input to show results
    sampleinput = (
    "HELLO, WORLD! \nTHE SUN RISES AT 6:30 AM. \n"
    "CHILDREN LAUGH & PLAY, \nWHILE THE BAKER PREPARES 50+ LOAVES OF BREAD. \n"
    "LIFE MOVES GENTLY; \nEACH DAY BRINGS NEW ADVENTURES. \n"
    "CAN YOU BELIEVE IT? YES, IT'S REAL! \nCONTACT US @BROOKFIELD.COM FOR UPDATES.\n"
   )


    image = rendertexttoimage(sampleinput, scalepix=8, charspace=1, linespace=2, paddingback=12)
    image.show()
    #saving results and printing them
    image.save("output.png")
    print("Saved output.png")
```
